```python
#Use OneHotEncoder to fill dummy data.
dummies = pd.get_dummies(df.town)
```

```python
dummies
```

```python
#Merge two dataframe - df & dummies
df_dummies = pd.concat([df, dummies], axis='columns')
```

```python
df_dummies
```

```python
# Drop one clumns out of 3 dummy variable columns due to Dummy Variable trap (the dummy variable trap is a problem that can occur when including all dummy variables representing categories of a categorical variable in a regression model, leading to multicollinearity. To avoid it, drop one level of the categorical variable when creating dummy variables.)
# Also drop "town" column as we don't have use in model as machine learning models work on numerical values.

```

```python
df_dummies.drop( [ 'town', 'west windsor'],axis='columns', inplace = True)
```

```python
df_dummies
```

```python
X = df_dummies.drop('price', axis='columns')
```

```python
y = df_dummies.price
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[3600,0,0]])
```

```python
model.score(X,y)
```
